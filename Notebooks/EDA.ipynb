# Initial Exploration of the Data

In this notebook, we embark on an initial exploration of a dataset composed of images. Our objective is to understand the composition of our dataset in terms of image categories. This exploration includes loading random images for visual inspection and quantitatively analyzing the dataset's structure.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from utils import load_random_images, count_items
```

## Loading and Displaying Random Images

To get a tangible sense of the images in our dataset, we start by loading and displaying random images from two categories: flipped and not flipped. This step is crucial for qualitative analysis, helping us verify the integrity of the dataset and the appropriateness of image categories.

```python
imgs = load_random_images(img_directory="./images/train/not_flip/",
                          n_images=5)

plt.figure(figsize=(20, 10))
for count, img in enumerate(imgs,
                            start=1):
    plt.subplot(1, 
                len(imgs),
                count)
    plt.imshow(img)
    plt.axis("off")
plt.show()
```

```python
imgs = load_random_images(img_directory="./images/train/flip/",
                          n_images=5)

plt.figure(figsize=(20, 10))
for count, img in enumerate(imgs,
                            start=1):
    plt.subplot(1, 
                len(imgs),
                count)
    plt.imshow(img)
    plt.axis("off")
plt.show()
```

## Dataset Structure Analysis

After visually inspecting individual images, the next step is to quantitatively analyze the structure of our dataset. Specifically, we aim to count the number of images in each category across our training and test sets. This analysis provides insights into the dataset's balance and informs decisions for model training and validation. We utilize a pandas DataFrame to aggregate these counts and calculate the proportions of each category within the training and test sets. This analysis is summarized in the table below, with proportions rounded to five decimal places for precision and clarity.

```python
df = pd.DataFrame(data={"nr_image_flip": [count_items(path="./images/train/flip/"), 
                                          count_items(path="./images/test/flip/")], 
                        "nr_image_not_flip": [count_items(path="./images/train/not_flip/"), 
                                              count_items(path="./images/test/not_flip/")]},
                  index=["train_data", 
                    "test_data"])

df.divide(other=df.sum(axis=1), axis=0).round(5)
```
